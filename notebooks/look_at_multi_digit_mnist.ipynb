```python
import random

%matplotlib inline
import matplotlib.pyplot as plt

from handwritten_digit_string_recognition.data import MultiDigitMNIST
```

```python
multi_digit_mnist = MultiDigitMNIST()
multi_digit_mnist.prepare_data()
multi_digit_mnist.setup()
```

```python
dataset = multi_digit_mnist.dataset["val"]
```

```python
for i in range(5):
    plt.figure(figsize=(8, 8))
    rand_idx = random.randint(0, len(dataset) - 1)
    image, label = dataset[rand_idx]
    plt.title(label)
    plt.imshow(image.squeeze(), cmap="gray")
```
